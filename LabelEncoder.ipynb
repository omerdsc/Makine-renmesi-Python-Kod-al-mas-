```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("Tour_Winners_data_1.csv")
df.head(2)
```

```python
le=LabelEncoder()
le.fit(df["Team"])
```

```python
x=le.transform(df["Team"])#veriyi numeriğe çevirdik
x
```

```python
le.classes_ #numeriğe çevirilen verileri bize verir
```

```python
le.inverse_transform(x) # nujmeric verilern isimlerini bize gösterir
```
